```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Fifa_world_cup_matches.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.hist(df['number of goals team1'])
plt.hist(df['number of goals team2'])
```

```python
goal_diff = df['number of goals team1'] - df['number of goals team2']
plt.hist(goal_diff)
```

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
sns.scatterplot(x='possession team1', y='number of goals team1', data=df)
```

```python
sns.distplot(df['yellow cards team1'])
sns.distplot(df['yellow cards team2'])
```

```python
df.plot.line(x='date', y=['red cards team1', 'red cards team2'])
```

```python
sns.barplot(x='penalties scored team1', y='penalties scored team2', data=df)
```

```python
pd.plotting.scatter_matrix(df[['possession team1', 'number of goals team1',
                               'total attempts team1', 'passes team1']])
```
